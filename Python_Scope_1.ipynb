# **Python scope defines the region of a program where a variable or name is visible and accessible.**

 It determines where a variable can be used and which parts of the code have access to it. Understanding scope is crucial for preventing naming conflicts, organizing code, and ensuring proper variable usage.
Python uses the LEGB rule (Local, Enclosing, Global, Built-in) to resolve names and determine their scope:
Local (L): This refers to the innermost scope, typically within a function. Variables defined inside a function are local to that function and cannot be accessed from outside it.

```python
def greet():

    # local variable
    message = 'Hello'

    print('Local', message)

greet()

# try to access message variable
# outside greet() function


# declare global variable
message = 'Hello'
print(message)

def greet():
    # declare local variable
    print('Local', message)

greet()
print('Global', message)


# outside function
def outer():
    message = 'local'

    # nested function
    def inner():

        # declare nonlocal variable
        nonlocal message

        message = 'nonlocal'
        print("inner:", message)

    inner()
    print("outer:", message)

outer()

# This function has a variable with
# name same as s.
def f():
    s = "Me too."
    print(s)

# Global scope
s = "I love Geeksforgeeks"
f()
print(s)

def f():
    print(s)

    # This program will NOT show error
    # if we comment below line.
    s = "Me too."

    print(s)


# Global scope
s = "I love Geeksforgeeks"


print(s)


# This function modifies global variable 's'
def f():
    global s
    print(s)
    s = "Look for Geeksforgeeks Python Section"
    print(s)


# Global Scope
s = "Python is great !"
f()
print(s)

# Python program to demonstrate
# scope of variable

a = 1

# Uses global because there is no local 'a'
def f():
    print('Inside f() : ', a)

# Variable 'a' is redefined as a local
def g():
    a = 2
    print('Inside g() : ', a)

# Uses global keyword to modify global 'a'
def h():
    global a
    a = 3
    print('Inside h() : ', a)

# Global scope
print('global : ', a)
f()
print('global : ', a)
g()
print('global : ', a)
h()
print('global : ', a)

# Python program to demonstrate
# nonlocal keyword

print("Value of a using nonlocal is : ", end="")


def outer():
    a = 5

    def inner():
        nonlocal a
        a = 10
    inner()
    print(a)


outer()

# demonstrating without non local
# inner loop not changing the value of outer a
# prints 5
print("Value of a without using nonlocal is : ", end="")


def outer():
    a = 5

    def inner():
        a = 10
    inner()
    print(a)


outer()

# Python program to demonstrate
# nonlocal keyword

print("Value of a using nonlocal is : ", end="")


def outer():
    a = 5

    def inner():
        nonlocal a
        a = 10
    inner()
    print(a)


outer()

# demonstrating without non local
# inner loop not changing the value of outer a
# prints 5
print("Value of a without using nonlocal is : ", end="")


def outer():
    a = 5

    def inner():
        a = 10
    inner()
    print(a)


outer()

def my_function():
    local_variable = 10
    print(local_variable)

my_function()
def modify_global():
    global global_variable
    global_variable = "Modified globally"

modify_global()
global_variable = "I am global"

def my_function():
    print(global_variable)

my_function()


print(global_variable)
def outer_function():
    outer_variable = "Original"

    def inner_function():
        nonlocal outer_variable
        outer_variable = "Modified inside"

    inner_function()
    print(outer_variable)

def outer_function():
    outer_variable = "I am outside"

    def inner_function():
        print(outer_variable)

    inner_function()

outer_function()

```
