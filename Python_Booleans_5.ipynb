# **Python Boolean**
In Python, the boolean datatype represents one of two built-in values: True or False. Booleans are essential for decision-making and control flow in programming.

```python
is_raining = True
print(is_raining)  # Output: True

is_sunny = False
print(is_sunny)  # Output: False

x = 5
y = 10

print(x == y)  # Output: False
print(x != y)  # Output: True
print(x < y)   # Output: True
print(x > y)   # Output: False
print(x <= y)  # Output: True
print(x >= y)  # Output: False

is_raining = True
is_cold = False

# Using 'and'
if is_raining and not is_cold:
	print('It is raining but not cold.') # Output: It's raining but not cold.

is_sunny = True

# Using 'or'
if is_raining or is_sunny:
	print('It is either raining or sunny.') # Output: It's either raining or sunny.


print(bool(0))        # Output: False
print(bool(42))       # Output: True
print(bool(''))       # Output: False
print(bool('Hello'))  # Output: True

print(10 > 9)
print(10 == 9)
print(10 < 9)

print(bool('Hi'))
print(bool(''))
print(bool(0))
print(bool(101))
print(bool(True))
print(bool(False))
print(bool(None))

is_python_fun = True
is_math_hard = False
print(is_python_fun)  # Output: True
print(is_math_hard)   # Output: False

x = 10
y = 20
print(x > y)   # Output: False
print(x == 10) # Output: True

if x == 10:  # SyntaxError
    print("x is 10")

x = 10
y = 20
print(x > 5 and y < 30)  # Output: True
print(x > 5 or y > 50)   # Output: True
print(not (x > 5))       # Output: False

x = 10
y = 20
print(x > 5 or y < 10 and y > 30)  # Unexpected result!

print(x > 5 or (y < 10 and y > 30))

print(bool(0))        # Output: False
print(bool(1))        # Output: True
print(bool(""))       # Output: False
print(bool("Python")) # Output: True
print(bool([]))       # Output: False
print(bool([1, 2, 3]))# Output: True

is_raining = True
if is_raining:
    print("Take an umbrella!")
else:
    print("Enjoy the sunshine!")

x = 5
y = 10
if x and y:  # Works but might be unclear
    print("Both x and y are truthy")

if x != 0 and y != 0:
    print("Both x and y are nonzero")

def expensive_function():
    print("Function executed!")
    return True

print(True or expensive_function())  # Output: True (function not executed)
print(False and expensive_function()) # Output: False (function not executed)

def is_even(number):
    return number % 2 == 0

print(is_even(10))  # Output: True
print(is_even(7))   # Output: False

print(True == 1)  # Output: True
print(False == 0)  # Output: True
print(True + True)  # Output: 2


print(bool(0))       # Output: False
print(bool(''))      # Output: False
print(bool([]))      # Output: False
print(bool(None))    # Output: False


print(bool(42))       # Output: True
print(bool('Python')) # Output: True
print(bool([1, 2, 3])) # Output: True


result = True and False
print(result)  # Output: False

result = True and bool(2)
print(result)  # Output: True


result = not True
print(result)  # Output: False

result = not False
print(result)  # Output: True

x = 10
y = 20

print(x < y)  # Output: True
print(x == y) # Output: False


x = 10
print(5 < x < 15)  # Output: True


age = 20

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")


def greet(name: str, excited: bool = False) -> str:
    if excited:
        return f"Hello, {name}!!!"
    return f"Hello, {name}."

print(greet("Alice"))         # Output: Hello, Alice.
print(greet("Alice", True))   # Output: Hello, Alice!!!


from typing import List

numbers: List[int] = [0, 1, 2, 3, 4, 5]
filtered = list(filter(lambda x: x % 2 == 0, numbers))

print(filtered)  # Output: [0, 2, 4]


def divide(a: int, b: int) -> float:
    assert b != 0, "Division by zero is not allowed."
    return a / b

print(divide(10, 2))  # Output: 5.0
# print(divide(10, 0))  # Raises AssertionError


print(True == 1)  # Output: True
print(True is 1)  # Output: False


print(bool([]))  # Output: False
```

# **Falsy Values**
The following values are considered falsy in Python:

None
False
0 (any numeric type: 0, 0.0, 0j, etc.)
Empty sequences or collections: '', (), [], {}, set()
Objects of custom classes with __bool__() or __len__() returning False
