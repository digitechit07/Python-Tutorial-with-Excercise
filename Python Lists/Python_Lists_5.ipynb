# **A Python lis**t is a built-in, ordered, mutable collection type that holds references to objects. Lists are one of Python’s most versatile container types and are used to group values (of any type) into a single variable while preserving order and allowing modification.

Key properties

Ordered: elements have a defined position (index). Indexing starts at 0; negative indices access from the end.
Mutable: you can change, add, or remove elements after creation.
Heterogeneous: elements can be of different types in the same list (ints, strings, objects, other lists).
Dynamic size: length can grow or shrink at runtime.
Elements are references: assignment stores references to objects; two lists can reference the same object.
Core operations and examples

Creation:
Literal: [1, 2, 3], ['a', 2, 3.0], [] (empty list)
Constructor: list(iterable) — e.g., list(range(5)) -> [0,1,2,3,4]
Indexing and slicing:
a[0], a[-1]
a[1:4], a[:], a[::-1] (slice produces new list)
Mutation:
Assignment: a[2] = 'x'
append/extend/insert: a.append(x), a.extend([y,z]), a.insert(1, val)
remove/pop/clear: a.remove(val), a.pop(), a.pop(index), a.clear()
Length and membership:
len(a), x in a, x not in a
Iteration:
for x in a: ...
enumerate(a) for index+value
Copying:
Shallow copy: b = a[:] or a.copy() (copies list object, not nested elements)
Deep copy: import copy; b = copy.deepcopy(a) (recursively copies nested objects)
Combining and repetition:
Concatenation: a + b -> new list
Repetition: a * 3 -> repeats elements
Sorting and reversing:
a.sort() sorts in place (optional key and reverse)
sorted(a) returns new sorted list
a.reverse() reverses in place
Performance characteristics (typical CPython)

Indexing and assignment by index: O(1)
Append amortized O(1)
Insert or delete near ends: O(1) amortized (append/pop); insert/pop at arbitrary index: O(n) due to shifting
Membership test (x in a): O(n)
Memory: contiguous array of pointers to PyObjects (overhead for resizing and object references)
When to use lists

When order matters and elements will be changed (added/removed).
When you need fast random access by integer index.
For small to medium collections; for very large collections or specific needs consider array.array, deque (collections.deque), tuple (immutable), set (unique unordered), dict (keyed lookup), or numpy arrays for numeric performance.
Common pitfalls

Mutable default arguments: def f(a=[]): ... reuses same list across calls; use None fallback.
Copy vs reference: b = a makes b reference same list; modifying b changes a.

# **What to Keep in Mind**
count() performs a linear search every time you call it. It’s fast enough for small lists, but for larger datasets, consider precomputing counts with Counter.
It only matches exact values. You won’t get partial matches unless you process the list first.
It’s case-sensitive for strings.
The list doesn’t get modified. count() is read-only and returns a new integer.

List Slicing
A slice, or sub-list of Python list elements can be selected from a list using a colon-separated starting and ending point.

The syntax pattern is myList[START_NUMBER:END_NUMBER]. The slice will include the START_NUMBER index, and everything until but excluding the END_NUMBER item.

When slicing a list, a new list is returned, so if the slice is saved and then altered, the original list remains the same.
