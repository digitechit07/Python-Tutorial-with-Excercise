We use them to solve simple or complex math problems while coding. They help us perform calculations directly in our programs.

```python
x = 10
y = 3

# Addition
print(x + y)

# Subtraction
print(x - y)

# Multiplication
print(x * y)

# Division
print(x / y)

# Floor Division
print(x // y)

# Modulus
print(x % y)

# Exponentiation
print(x ** y)

result = 5 + 2 * 3 ** 2
print(result)  # Output: 23

result = 5 + 2 * 3 ** 2
print(result)  # Output: 23

x = 10
y = 3

# Assignment
x = y + 2
print("x = y + 2:", x)

# Addition Assignment
x += y  # x = x + y
print("x += y:", x)

# Subtraction Assignment
x -= y  # x = x - y
print("x -= y:", x)

# Multiplication Assignment
x *= y  # x = x * y
print("x *= y:", x)

# Division Assignment
x /= y  # x = x / y
print("x /= y:", x)

# Remainder Assignment
x %= y  # x = x % y
print("x %= y:", x)

# Floor Division Assignment
x //= y  # x = x // y
print("x //= y:", x)

# Exponent Assignment
x **= y  # x = x ** y
print("x **= y:", x)

x = True
y = False

# Logical AND
print("x and y:", x and y)

# Logical OR
print("x or y:", x or y)

# Logical NOT
print("not x:", not x)
print("not y:", not y)

x = 6   # binary: 110
y = 3   # binary: 011

# Bitwise AND
print("x & y:", x & y)

# Bitwise OR
print("x | y:", x | y)

# Bitwise XOR
print("x ^ y:", x ^ y)

# Bitwise NOT
print("~x:", ~x)

# Bitwise Right Shift
print("x >> 1:", x >> 1)

# Bitwise Left Shift
print("x << 1:", x << 1)

x = [10, 20]
y = x
z = [10, 20]

# Identity Operator: is
print("x is y:", x is y)     #  y points to the same object as x

# Identity Operator: is not
print("x is z:", x is z)     #  z has same values but is a different object
print("x is not z:", x is not z)  #  x and z are not the same object

fruits = ["apple", "banana", "mango"]

# Using 'in' operator
print("apple in fruits:", "apple" in fruits)

# Using 'not in' operator
print("grape not in fruits:", "grape" not in fruits)

x = 10
y = 20

# Using ternary operator to find the greater value
max_value = x if x > y else y
print("Greater value is:", max_value)

# Another example with a condition
status = "Pass" if x >= 10 else "Fail"
print("Status:", status)

```

# **Python Operators**
Python has different types of operators for different operations. They are as follows:

Arithmetic operators:
Arithmetic operators are used to perform arithmetic/mathematical operations.

Name	Operator	Example
Addition	+	a+b

Subtraction	-	a-b

Multiplication	*	a*b

Division	/	a/b

Exponential	**	a**b

Modulus	%	a%b

Floor Division	//	a//b

Assignment operators:
These operators are used to assign values to variables.

Name	Evaluated As
=	a=b

+=	a+=b or a=a+b

-=	a-=b or a=a-b

*=	a*=b or a=a*b

**=	a**=b or a=a**b

/=	a/=b or a=a/b

//=	a//=b or a=a//b

%=	a%=b or a=a%b

&=	a&=b or a=a&b

**	=**

^=	a^=b or a=a^b

=	a>>=b or a=a>>b

<<=	a<<=b or a=a<<b

```python
# Initial assignment
x = 10
print(x)
# output: 10

# Augmented assignment operator: subtraction
x -= 2
print(x)
# output: 8

# Augmented assignment operator: multiplication
x *= 3
print(x)
# output: 24

option = input("Please select option 1 or 2: ")
if option == "1":
    print("Option 1 was selected")
elif option == "2":
    print("Option 2 was selected")
else:
    print("Invalid option")

sun_is_shining = True
have_to_work = False

print("Should I go to the park?")
answer = sun_is_shining and not have_to_work

if answer == True:
    print("Yes!")
else:
    print("No!")


list_a = [1, 2, 3]
list_b = [1, 2, 3]

print(list_a == list_b)  # output: True
print(list_a is list_b)  # output: False

# Modifying list_a does not change list_b
list_a.clear()

print(list_a)  # output: []
print(list_b)  # output: [1, 2, 3]


x = 10
y = 5

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y)
print(x ** y)
print(x // y)

x = 10
y = 5

# Simple assignment
x = y
print(x)

# Addition assignment
x += y
print(x)

# Subtraction assignment
x -= y
print(x)

# Multiplication assignment
x *= y
print(x)

# Division assignment
x /= y
print(x)

# Modulus assignment
x %= y
print(x)

# Exponentiation assignment
x **= y
print(x)

# Floor division assignment
x //= y
print(x)

x = 5
y = 10

print(x == y)

print(x != y)

print(x > y)

print(x <= y)

a = True
b = False

# Check if a and b are both True
print(a and b)

# Check if a or b is True
print(a or b)

# Check if a is not True
print(not a)

x = [1, 2, 3]
y = [1, 2, 3]
z = x

# Check if x and y are the same object
print(x == y)  # False

# Check if x and z are the same object
print(x is z)  # True

# Check if x and y are not the same object
print(x is not y)  # True

x = [1, 2, 3]
y = 2

# Check if y is in x
print(y in x)

# Check if y is not in x
print(y not in x)

p = 5  # 101
q = 3  # 011

# AND
print(p & q)

# OR
print(p | q)

# XOR
print(p ^ q)

# NOT
print(~p)

# left shift
print(p << 1)

# right shift
print(p >> 1)

# Precedence-arithmetic operators
x = 2 + 3 * 4
y = (2 + 3) * 4
print(x,",",y)

# Precedence-bitwise and logical operators
x = True and False or True
y = True or False and False
print(x,",",y)

# Associativity-exponentiation operator
x = 2 ** 3 ** 2
y = (2 ** 3) ** 2
print(x,",",y)

# Associativity-assignment operator
x = y = z = 0   # x, y, and z are all set to 0
print(x,",",y)

#dividor operator
x = 123/5
y = 123//5
print(x,",", y)

x = 5
y = 10
max_val = x if x > y else y
print(max_val)

x = 5
y = 10
print("x is larger" if x > y else "y is larger")

m = 5
n = 10
ans = (m, n) if m > n else (n, m)
print(ans)


p = 12
q = 10
res = {True: p, False: q}[p > q]
print(res)

x = 5
y = 10
result = (lambda: x, lambda: y)[x > y]()
print(result)

p = 5
q = 10
r = 15
result = p if p > q else q if q > r else r
print(result)

x = 5
y = 10
print("x is larger" if x > y else "y is larger")

x = 15
y = 10
result = "x is larger" if x > y else "y is larger"
print(result)


```
